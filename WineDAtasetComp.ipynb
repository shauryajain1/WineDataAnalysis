```python
from pycaret.datasets import get_data
dataSets = get_data('index')
```

```python
wineDataSet = get_data("wine")
```

```python
from pycaret.classification import *
s = setup(data=wineDataSet, target='quality')
```

```python
cm = compare_models()
```

```python
setup(data=wineDataSet, target='quality',
      pca = True, pca_method = 'linear')
cm = compare_models()
```

```python
rfModel = create_model('rf')
```

```python
newDataSet = get_data("wine").iloc[:500]
```

```python
newPredictions = predict_model(rfModel, data = newDataSet)
newPredictions
```
